```python
import pandas as pd

# Load the proposal data
proposal = pd.read_excel(r'D:\DAO_Voting_1101\proposals_snapshot.xlsx')

# Filter proposals with more than 10 votes
filtered_proposals = proposal[proposal['votes'] > 10]

# Get the number of unique 'id' values that have more than 10 votes
num_ids_with_more_than_10_votes = filtered_proposals['id'].nunique()

# Find the oldest and latest dates from the 'created' column
oldest_date = filtered_proposals['created'].min()
latest_date = filtered_proposals['created'].max()

# Calculate the average and sum of votes after filtering
average_votes = filtered_proposals['votes'].mean()
sum_votes = filtered_proposals['votes'].sum()

# Get the unique vote of 'space_id' column
unique_space_ids = filtered_proposals['space_id'].nunique()

# Display the results
print("Number of unique 'id' with more than 10 votes:", num_ids_with_more_than_10_votes)
print("Oldest date:", oldest_date)
print("Latest date:", latest_date)
print("Average votes after filtering:", average_votes)
print("Sum of votes after filtering:", sum_votes)
print("Unique number of 'space_id':", unique_space_ids)
```

```python
#
import pandas as pd
import os

# Directory containing the CSV files
directory = r'D:\DAO_Voting_1101\Statistics\voter_stats\tipping_point\dynamic_pivotal_voters\by_protocol\pivotal_score_added'

# List to store DataFrames
dfs = []

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory, filename)
        df = pd.read_csv(file_path)
        dfs.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(dfs, ignore_index=True)

# Display the combined DataFrame
print(combined_df)
```

```python
combined_df.to_csv(r'D:\DAO_Voting_1101\Statistics\voter_stats\tipping_point\dynamic_pivotal_voters\by_protocol\combined_data.csv', index=False)
```
