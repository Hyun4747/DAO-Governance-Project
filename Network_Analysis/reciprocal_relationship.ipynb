```python
import os
import pandas as pd
from tqdm import tqdm

def find_reciprocal_relationships(df):
    reciprocal_relationships = []

    # Extract relevant columns
    df['proposal_ids'] = df['proposal_ids'].apply(eval)
    df['author'] = df['author'].apply(eval)
    df['choice'] = df['choice'].apply(eval)

    print(df.head())
    output_directory = r"D:\DAO_Voting_1101\Statistics\voter_stats\reciprocity"
    df.to_csv(os.path.join(output_directory, 'df.csv'))

    # Create a cartesian product of rows for comparison
    cartesian_product = pd.merge(df, df, how='cross')

    print(cartesian_product.head())
    cartesian_product.to_csv(os.path.join(output_directory, 'cartesian_product.csv'))

    # Filter pairs where voter and author are reversed
    reciprocal_pairs = cartesian_product[
        (cartesian_product['voter_address_x'] == cartesian_product['author_y']) &
        (cartesian_product['author_x'] == cartesian_product['voter_address_y'])
    ]

    # Extract the relevant information
    reciprocal_relationships = reciprocal_pairs[[
        'voter_address_x', 'proposal_ids_x', 'author_x', 'choice_x'
    ]].rename(columns={
        'voter_address_x': 'voter_address',
        'proposal_ids_x': 'proposal_id',
        'author_x': 'author',
        'choice_x': 'choice'
    })

    return reciprocal_relationships
```

```python
# Define the directory path
directory_path = r"D:\DAO_Voting_1101\Statistics\voter_stats\multiple_times_voters"

# Output directory for reciprocal relationships
output_directory = r"D:\DAO_Voting_1101\Statistics\voter_stats\reciprocity"

# Iterate through each CSV file in the directory
for filename in os.listdir(directory_path)[:1]:
    if filename.endswith(".csv"):
        # Read the CSV file
        file_path = os.path.join(directory_path, filename)
        df = pd.read_csv(file_path)

        # Find Reciprocal Relationships
        reciprocal_relationships = find_reciprocal_relationships(df)

        # Export to CSV
        reciprocal_relationships.to_csv(os.path.join(output_directory, f"{filename.split('.')[0].split('_')[0]}_reciprocal_relationships.csv"), index=False)
        print(f"Exported {filename} to CSV")
```
